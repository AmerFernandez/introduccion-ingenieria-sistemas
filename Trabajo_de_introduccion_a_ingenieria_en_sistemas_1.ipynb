```python
# Ejercicio numero 4
edadjuan = 9
edadalberto = (2/3) * edadjuan
edadana = (4/3) * edadjuan
edadmama = edadjuan+edadalberto+edadana
print("Edad de Juan ", edadjuan)
print("Edad de Alberto ", edadalberto)
print("Edad de Ana", edadana)
print("Edad de Mama", edadmama)
```

```python
# Ejercicio numero 12
nombre_empleado = input("Nombre del empleado: ")
horas_de_trabajo = float(input("Horas de trabajo: "))
salario_por_hora = 5000
print("Salario por hora: ", salario_por_hora)
salario_bruto = horas_de_trabajo*salario_por_hora
print("El salario bruto es: ", salario_bruto)
rentencion_fuente = salario_bruto*0.125
print("La rentencion de la fuente es: ", rentencion_fuente)
salario_neto = salario_bruto-rentencion_fuente
print("El salario neto es: ", salario_neto)


```

```python
# Ejercicio numero 14
ingrese_numero = float(input("Ingrese su número: "))
numero_cuadrado = ingrese_numero**2
print("Su número elevado al cuadrado es: ", numero_cuadrado)
numero_cubo = ingrese_numero**3
print("Su número elevado al cubo es: ", numero_cubo)
```

```python
# Ejercicio numero 17
numero_radio = float(input("Ingrese el radio del circulo: "))
pi = 3.1416
area = pi * (numero_radio ** 2)
print("El area del circulo es: ", area)
longitud = 2 * pi * numero_radio
print("La longitud de la circunferencia es: ", longitud)
```
